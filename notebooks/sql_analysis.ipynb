# Vivino Market Analysis Notebook

This notebook connects to the SQLite database, loads each SQL analysis file, runs the query, and displays the result.

```python
from pathlib import Path
import sqlite3
import pandas as pd

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
db_path = project_root / 'data' / 'vivino.db'
sql_dir = project_root / 'sql'

print('Project root:', project_root)
print('Database path:', db_path)
print('SQL directory:', sql_dir)
```

```python
conn = sqlite3.connect(db_path)
conn
```

```python
analysis_files = [
    'top_10_wines.sql',
    'best_countries.sql',
    'winery_awards.sql',
    'keyword_analysis.sql',
    'grape_analysis.sql',
    'country_leaderboard.sql',
    'vintage_leaderboard.sql',
]

analysis_files
```

```python
def run_sql_file(filename: str) -> pd.DataFrame:
    query = (sql_dir / filename).read_text(encoding='utf-8')
    print(f'Running: {filename}')
    return pd.read_sql_query(query, conn)
```

```python
results = {}
for filename in analysis_files:
    results[filename] = run_sql_file(filename)

list(results.keys())
```
